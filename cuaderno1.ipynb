```python
lista = []
```

```python
for x in range(0,1000):
    lista.append({'nombre': fake.name(),'apellido': fake.last_name(),'correo': fake.email(),'dirección': fake.address()})
```

```python
import pandas as pd
data = pd.DataFrame(lista)
```

```python
import random
n = len(data)
provincias = ['Loja', 'Pichincha', 'Guayas', 'Azuay', 'Manabí']
data['edad']= [random.randint(30, 40) for _ in range(n)]
data['provincias'] = [random.choice(provincias) for _ in range(n)]
data
```

```python
data.to_csv("cuaderno1.csv", index=False)
```
